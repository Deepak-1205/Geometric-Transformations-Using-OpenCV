# **Image Translation**

Write a Python program using OpenCV to perform image translation on the **"Given image"**.

The program should:

1. Import `cv2`, `numpy`, and `matplotlib.pyplot`.
2. Read the image **"given** in color mode.
3. Create a **translation matrix** that moves the image  
   **50 pixels to the right** and **80 pixels down**.
4. Apply the translation
5. Display the following using a **1 × 2 layout** in Matplotlib:  
   - **Original Image**  
   - **Translated Image**


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('baseball.jpg')
```

```python
image.shape
```

```python
plt.imshow(image[:,:,::-1])
plt.title("Original Image")
plt.show()
```

```python
tx,ty=50,80
M=np.float32([[1,0,tx],[0,1,ty]])
h,w=image.shape[:2]
translated_image = cv2.warpAffine(image, M, (w,h))
```

```python
plt.imshow(translated_image[:,:,::-1])
plt.title("Translated Image")
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1);plt.imshow(image);plt.title("Original Image")
plt.subplot(1,2,2);plt.imshow(translated_image);plt.title("Translated Image")
```

# **Image Shearing**

Write a Python program using OpenCV to perform image shearing on the **given image**.

The program should:

1. Create a **shearing transformation matrix** to apply:
   - **Horizontal shearing**  
   - **Vertical shearing**
2. Apply the shearing transformations.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Sheared Image**  
   - **Vertically Sheared Image**


```python
sh_x = 0.3
sh_y = 0.2
```

```python
M_horizontal = np.float32([[1, sh_x, 0], [0, 1,    0]])
M_vertical   = np.float32([[1,    0,    0], [sh_y, 1,    0]])
```

```python
h, w = image.shape[:2]
```

```python
shear_horiz = cv2.warpAffine(image, M_horizontal, (int(w + h * sh_x), h))
shear_vert  = cv2.warpAffine(image, M_vertical, (w, int(h + w * sh_y)))
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 3, 1); plt.imshow(image[:, :, ::-1]); plt.title("Original Image")
plt.subplot(1, 3, 2); plt.imshow(shear_horiz[:, :, ::-1]); plt.title("Horizontally Sheared")
plt.subplot(1, 3, 3); plt.imshow(shear_vert[:, :, ::-1]); plt.title("Vertically Sheared")
```

# **Image Rotation**

Write a Python program using OpenCV to perform **image rotation** on the **given image**.

The program should:

1. Rotate the image by:  
   - **45 degrees**  
   - **90 degrees**
2. Use `cv2.getRotationMatrix2D` and `cv2.warpAffine`.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **45° Rotated Image**  
   - **90° Rotated Image**


```python
(h,w)=image.shape[:2]
center = (w // 2, h // 2)
```

```python
M_45 = cv2.getRotationMatrix2D(center, 45, 1)
M_90 = cv2.getRotationMatrix2D(center, 90, 1)
```

```python
rotated_45 = cv2.warpAffine(image, M_45, (w, h))
rotated_90 = cv2.warpAffine(image, M_90, (w, h))
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 3, 1); plt.imshow(image[:, :, ::-1]); plt.title("Original Image")
plt.subplot(1, 3, 2); plt.imshow(rotated_45[:, :, ::-1]); plt.title("45° Rotated Image")
plt.subplot(1, 3, 3); plt.imshow(rotated_90[:, :, ::-1]); plt.title("90° Rotated Image")
```
